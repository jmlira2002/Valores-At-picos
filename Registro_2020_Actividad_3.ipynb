```python
#Importamos la libreria pandas y cargamos archivo excel
import pandas as pd
df = pd.read_excel('/content/Gastos y costos 20-23.xlsx', sheet_name='2020')
df
```

```python
#Eliminamos las filas incesarias, ya que las primeras 5 no contienen información
df1 = pd.read_excel('/content/Gastos y costos 20-23.xlsx', sheet_name='2020', skiprows=5)
df1
```

```python
# Encontramos que tenemos un total de 3342 registros y 19 columnas
df1.shape
```

```python
# Contanmos el total de nulos por columna
df1.isnull().sum()
```

```python
# Encontramos que la columna 'Poliza' cast totalmente vacia, por lo que
# Consideramos eliminarla, ya que de 3342 registros, solo 21 han sido
# caapturados
df1.drop('POLIZA', axis=1, inplace=True)
# Corroboramos que se haya eliminado
df1.isnull().sum()
```

```python
# En la columna folio hemos sustituido los valores nulos por los valores no nulos
# hacia delante "Foward fill"
df1["FOLIO"] = df1["FOLIO"].fillna(method="ffill")
```

```python
# corroboramos 0 nulos en columna folio
df1.isnull().sum()
```

```python
# Al ser categorica, buscamos la moda con value_counts() en la columna Gasto
df1['GASTO'].value_counts()
```

```python
# Sustituimos los valores nulos con la moda, en este caso por "COMPRA"
df1["GASTO"] = df1["GASTO"].fillna("COMPRA")
```

```python
#Corroboramos no nulos en la columna gasto
df1.isnull().sum()
```

```python
# Buscamos la moda en columna tipo para reemplazar el dato nulo
df1['TIPO'].value_counts()
```

```python
# Sustituimos todos el valor nulo por 'I'
df1["TIPO"] = df1["TIPO"].fillna('I')
```

```python
#Corroboramos no nulos en la columna Tipo
df1.isnull().sum()
```

```python
# Sustiuimos los no nulos de la columna TC con el promedio
df1["TC"]= df1["TC"].fillna(round(df1["TC"].mean(),1))
```

```python
#Corroboramos que no existan nulos la columna TC
df1.isnull().sum()
```

```python
# Sustiuimos los no nulos de la columna importe con la mediana
df1["IMPORTE"]= df1["IMPORTE"].fillna(round(df1["IMPORTE"].median(),1))
```

```python
#Corroboramos que no existan nulos la columna importe
df1.isnull().sum()
```

```python
# Sustiuimos los no nulos de la columna IVA con el promedio
df1["IVA"]= df1["IVA"].fillna(round(df1["IVA"].mean(),1))
```

```python
#Analizamos que no exista ningun nulo mas en el df
df1.isnull().sum()
```

```python
#Realizamos diagrama de caja del df
import matplotlib.pyplot as plt
fig = plt.figure(figsize =(15, 8))
df1.plot(kind='box', vert=False)
plt.title("Valores Atípicos")
plt.show()
```

```python
# LIMITES DE TOTAL SAT
Limite_Superior_TOTAL_SAT = df1['TOTAL SAT'].mean() + 3 * df1['TOTAL SAT'].std()
Limite_Inferior_TOTAL_SAT = df1['TOTAL SAT'].mean() - 3 * df1['TOTAL SAT'].std()
# LIMITES DE TOTAL MX
Limite_Superior_TOTAL_MX = df1['TOTAL MX'].mean() + 3 * df1['TOTAL MX'].std()
Limite_Inferior_TOTAL_MX = df1['TOTAL MX'].mean() - 3 * df1['TOTAL MX'].std()
# LIMITES DE RET IVA
Limite_Superior_RET_IVA = df1['RET IVA'].mean() + 3 * df1['RET IVA'].std()
Limite_Inferior_RET_IVA = df1['RET IVA'].mean() - 3 * df1['RET IVA'].std()
# LIMITES DE RET ISR
Limite_Superior_RET_ISR = df1['RET ISR'].mean() + 3 * df1['RET ISR'].std()
Limite_Inferior_RET_ISR = df1['RET ISR'].mean() - 3 * df1['RET ISR'].std()
```

```python
df1['TOTAL SAT'] = df1['TOTAL SAT'].apply(lambda x: x if (Limite_Inferior_TOTAL_SAT <= x) & (x <= Limite_Superior_TOTAL_SAT) else None)
df1['TOTAL MX'] = df1['TOTAL MX'].apply(lambda x: x if (Limite_Inferior_TOTAL_MX <= x) & (x <= Limite_Superior_TOTAL_MX) else None)
df1['RET IVA'] = df1['RET IVA'].apply(lambda x: x if (Limite_Inferior_RET_IVA <= x) & (x <= Limite_Superior_RET_IVA) else None)
df1['RET ISR'] = df1['RET ISR'].apply(lambda x: x if (Limite_Inferior_RET_ISR <= x) & (x <= Limite_Superior_RET_ISR) else None)
```

```python
# Se han eliminado los outliers de las columnas total sat, total mx, ret iva, ret isr
df1.isnull().sum()
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
df1=df1.fillna(round(df1.mean(),1))
# Verificamos que no haya mas nulos
df1.isnull().sum()
```

```python
# IVA, IPORTE, TC y columna 1 SON LOS ULTIMOS EN CONTENER OUTLIERS, POR LO QUE LOS SUSTITUIMOS
# EL METODO DE LOS CUANTILES
columnas_interes = ['IVA', 'IMPORTE', 'TC', 'Columna1']

percentile25 = df1[columnas_interes].quantile(0.25)
percentile75 = df1[columnas_interes].quantile(0.75)

iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5 * iqr
Limite_Inferior_iqr = percentile25 - 1.5 * iqr

print("Límite superior permitido (IQR):")
print(Limite_Superior_iqr)

print("Límite inferior permitido (IQR):")
print(Limite_Inferior_iqr)
```

```python
# Eliminamos los outliers de las columnas IVA, IPORTE, TC y columna 1
for columna in columnas_interes:
    df1[columna] = df1[columna].apply(lambda x: x if (Limite_Inferior_iqr[columna] <= x) & (x <= Limite_Superior_iqr[columna]) else None)
```

```python
df1.isnull().sum()
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con el promedio
df1=df1.fillna(round(df1.mean(),1))
# Verificamos que no existan más nulos
df1.isnull().sum()
```
